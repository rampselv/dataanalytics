```python
#NORMAL DISTRIBUTION
import numpy as np 
import matplotlib.pyplot as plt

values= np.random.normal(200,20,20000)
print(values.shape)
plt.hist(values,100,density=True)
plt.show()
```

```python
#UNIFORM DISTRIBUTION
s = np.random.uniform(5,50,1000)
count, bins, ignored = plt.hist(s, 50,density=True)
plt.show()
```

```python
#POISSON DISTRIBUTION
pois = np.random.poisson(7.7, 7000)
plt.hist(pois, 10,density=True)
plt.show()
```

```python
#BINOMIAL DISTRIBUTION
n=100
p=0.2667
binom = np.random.binomial(n, p, 10000)
plt.hist(binom,25,density=True)
plt.show()
```

```python
#EXPONENTIAL DISTRIBUTION
target = 200
beta = 1.0/target
Y = np.random.exponential(beta, 3000)
plt.hist(Y, bins=150,lw=0,alpha=.7)
plt.plot([0,max(Y)],[target,target],'r--')
plt.ylim(0,target*1.1)
plt.show()
```

```python
#RANDOM DISTRIBUTION
ran=np.random.choice(range(5000,10000),100)
ran
plt.hist(ran,5)
plt.show()
```
